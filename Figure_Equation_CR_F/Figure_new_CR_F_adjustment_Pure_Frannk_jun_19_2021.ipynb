```
Num=7

for i in range(Num):
  plt.subplot(2,4,i+1)
  file_selected=Files[i]
  df=pd.read_excel(file_selected)
  Y_CR=np.array(df['F'])
  Y_F=np.array(df['CR'])
  Num_kk = len(Y_F)
  X_F = []
  Y_CR_new = []
  Y_F_new = []
  for kk, v in enumerate(Y_CR):
    if(kk % 80 == 0):
      Y_CR_new.append(Y_CR[kk])
      Y_F_new.append(Y_F[kk])
      X_F.append(kk/Num_kk)

  plt.xlim(0, 1.0)
  plt.ylim(-0.1, 0.6)
  plt.setp(ax1.spines.values(), linewidth=2.0)
  plt.plot(X_F, Y_F_new, 'ko-', X_F, Y_CR_new, 'k^--', linewidth=2, markersize=6)
  plt.legend( ('F', 'CR'), loc=[0.6,0.7], fontsize=12)
  # plt.text(-0.2,1.6,Legends[i],fontsize=16)
  plt.xlabel('Fraction iterations', fontsize=14, fontweight='bold')
  label_yy = Labels_yy[i] + 'F and CR'
  plt.ylabel(label_yy, fontsize=14, fontweight='bold')
  plt.grid(True, linewidth=1,color='gray')
  plt.tick_params(axis='both', which='major', labelsize=15)

plt.subplots_adjust(bottom=0.9, right=3, top=3, hspace=0.2, wspace=0.4)

plt.savefig('functions_CR_F.png', dpi=500,bbox_inches='tight')

CR_F = 'functions_CR_F.png'
```
