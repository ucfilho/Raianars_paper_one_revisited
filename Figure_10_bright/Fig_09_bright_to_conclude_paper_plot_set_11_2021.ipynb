```python
Files = ['SeiCo_Pure_Frannk.xlsx','Sphere_Pure_Frannk.xlsx','Rastrigin_Pure_Frannk.xlsx',
         'Schwefel_Pure_Frannk.xlsx','Sum_of_different_powers_Pure_Frannk.xlsx',
         'Ackley_Pure_Frannk.xlsx','Michalewicz_Pure_Frannk.xlsx']

Labels_yy = ['SeiCo:  ','Sphere: ','Rastrigin: ',
         'Schwefel 2.26: ','Sum of dif powers: ',
         'Ackley: ','Michalewicz: ']
```

```python
Num=7
ky = 0
vai = 0
foi = 0
nrows = Num

plt.figure(figsize=(20, 30))
vai = 0
foi = 0
fui = 0

for i in range(Num):
  ky = ky + 1
  file_selected=Files[i]
  df=pd.read_excel(file_selected)
  Y_CR=np.array(df['F'])
  Y_F=np.array(df['CR'])
  (rows,) = Y_CR.shape
  Num_kk = len(Y_F)

  X_F = []; Y_CR_new = []; Y_F_new = []; Y_CR1_new = []; Y_F1_new = []; X_F1 = []
  for kk in range(rows):
    if(kk % 10 == 0):
      Y_CR_new.append(Y_CR[kk])
      Y_F_new.append(Y_F[kk])
      X_F.append((kk+1)/rows)

    if(kk < 400 ):
      if(kk % 2 == 0):
        Y_CR1_new.append(Y_CR[kk])
        Y_F1_new.append(Y_F[kk])
        X_F1.append((kk+1)/rows)
  # print(X_F)
  vai = vai + 3
  
  plt.subplot(nrows, 3, vai-2)
  name = Labels_yy[i] + 'F and CR'
  plt.ylim(-0.2, 0.6)
  
  plt.plot(X_F, Y_F_new, 'k-', X_F, Y_CR_new, 'r-', linewidth=1, markersize=6)
  plt.legend( ('CR', 'F'), loc=[0.7,0.7], fontsize=10)
  plt.xlabel('Fraction iterations', fontsize=11, fontweight='bold')
  plt.ylabel(name, fontsize=11, fontweight='bold')
  plt.subplot(nrows, 3, vai-1)
  name = Labels_yy[i] + 'F (zoom first iterations)'
  plt.plot(X_F1, Y_CR1_new,'r', linewidth=1)
  plt.xlabel('Fraction iterations', fontsize=11, fontweight='bold')
  plt.ylabel(name, fontsize=11, fontweight='bold')
  plt.subplot(nrows, 3, vai)
  name = Labels_yy[i] + 'CR (zoom first iterations)'
  plt.plot(X_F1, Y_F1_new,'k', linewidth=1)
  plt.xlabel('Fraction iterations', fontsize=11, fontweight='bold')
  plt.ylabel(name, fontsize=11, fontweight='bold')

 
```
