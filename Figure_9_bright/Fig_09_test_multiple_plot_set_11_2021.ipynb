```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

N = 100
x1 = range(N)
y1 = np.cos(x1)
x2 = [2*i for i in x1]
y2 = y1**2
vai = 0
foi = 0
nrows = 10

plt.figure(figsize=(20, 90))

for i in range(nrows):
  vai = foi +1
  foi = vai + 1
  plt.subplot(nrows, 2, vai)
  plt.plot(x1, y1, 'tab:orange')
  plt.xlabel('x left')
  plt.subplot(nrows, 2, foi)
  plt.plot(x2, -y2, 'tab:green')
  plt.xlabel('x right')


```
